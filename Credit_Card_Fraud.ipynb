```python
for col in credit_card_data.columns:
    if credit_card_data[col].dtype != 'object':
        credit_card_data[col] = credit_card_data[col].fillna(credit_card_data[col].median())
```

```python
X= credit_card_data.drop(columns= 'Class', axis=1)
```

```python
Y = credit_card_data['Class']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, stratify=Y, random_state=2)
```

```python
smote = SMOTE(random_state=1)
```

```python
X_train_res, Y_train_res = smote.fit_resample(X_train, Y_train)
```

```python
print(Y_train_res.shape)
```

```python
X_train.head()
```

```python
print(X_train_res.shape)
```

```python
LORmodel = LogisticRegression(max_iter = 1000)
```

```python
LORmodel.fit(X_train_res,Y_train_res)
```

```python
X_train_prediction = LORmodel.predict(X_train_res)
```

```python
training_data_accuracy= accuracy_score(X_train_prediction, Y_train_res)
```

```python
print(training_data_accuracy)
```

```python
X_test_prediction = LORmodel.predict(X_test)
```

```python
print(accuracy_score(X_test_prediction, Y_test))
```
